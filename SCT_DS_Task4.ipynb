```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv"

df = pd.read_csv(url)

df.head()
```

```python
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nDataset Information:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
top = df.sort_values("total exports", ascending=False).head(10)

plt.figure(figsize=(10,6))

plt.bar(top["state"], top["total exports"])

plt.xticks(rotation=45)

plt.title("Top 10 States")

plt.xlabel("State")

plt.ylabel("Total Exports")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["total exports"], bins=20)

plt.title("Distribution")

plt.xlabel("Total Exports")

plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(df["corn"], df["total exports"])

plt.xlabel("Corn")

plt.ylabel("Total Exports")

plt.title("Corn vs Total Exports")

plt.show()
```

```python
corr = df.select_dtypes(include=np.number).corr()

plt.figure(figsize=(8,6))

plt.imshow(corr, cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")

plt.show()
```
